# Data Cleaning

## First Copy



*   The third row is just metadata that we can get rid of.
*   The first few columns also contains unnecessary survey data that we can get rid of.

```python
# Drop the second row (which contains metadata)
df_c1 = df.drop(1)
# Reset the index for good measure
df_c1 = df_c1.reset_index(drop=True)
# Drop all the columns before the first survey question
df_c1 = df_c1.iloc[:, 10:]

print(df_c1.shape)
# Display the first few rows after cleaning
df_c1.head()
```

## Seperate dataframe for mapping the sections and questions
Create a seperate dataframe that maps each question (e.g.,
headers to the section. E.g., "Have you used tinder duriong your travels => 1")

```python
# Extracting the headers (questions) to create a mapping
question_mapping = {question: idx + 1 for idx, question in enumerate(df_c1.columns)}

# Creating a DataFrame to represent the mapping
question_mapping_df = pd.DataFrame(list(question_mapping.items()), columns=['Question', 'Section'])

# Function to determine the section based on the question code, handling different formats
def get_section(question_code):
    # Extracting the first numeric portion
    section_number = ''.join(filter(str.isdigit, str(question_code).split('_')[0].split('.')[0]))
    return int(section_number) if section_number else None

# Applying the function to determine the sections
question_mapping_df['Section'] = question_mapping_df['Question'].apply(get_section)

# Displaying the updated mapping
question_mapping_df.head()
```

- Append the actual questions to the dataframe.

```python
questions_string = df_c1.iloc[0].values
# questions_string
question_mapping_df['String'] = questions_string
question_mapping_df
```

## Change questions as the header
- now that we have a specific section to map the questions for reference, we can continue with our data cleaning.


```python
# Drop the first row
df_2 = df_c1.drop(0)

# Set the next row as header
df_2.columns = df_c1.iloc[0]

# Reset index
df_2.reset_index(drop=True, inplace=True)

# Displaying the first few rows after making the changes
print(df_2.shape)
df_2.head()
```

## Second Copy - Taking only the number
- Since the survey used Likert Scale, we will change the answers with only the number of the selection


```python
# Function to extract numerical value from a string
def extract_number(value):
    if pd.isnull(value) or isinstance(value, (int, float)):
        return value
    match = re.match(r'^(\d+)', str(value))
    return int(match.group(1)) if match else value

# Applying the function to all elements in the DataFrame
df_c2 = df_2.applymap(extract_number)

# Displaying the first few rows after transformation
df_c2.head()
```

```python
df_c2.shape
```

## Third Copy
### 1. All yes/no answers to boolean


```python
# Function to convert "Yes" and "No" to binary values
def convert_yes_no(value):
    if value == "Yes":
        return True
    elif value == "No":
        return False
    return value

# Applying the function to all elements in the DataFrame
df_c3= df_c2.applymap(convert_yes_no)

# Displaying the first few rows after transformation
df_c3.head()
```

```python
df_c3.shape
```

### 2. All Male, Female turn into 0 & 1

```python
# Function to convert "Female" and "Male" to binary values
def convert_gender(value):
    if value == "Female":
        return 'F'
    elif value == "Male":
        return 'M'
    return value

# Applying the function to the specific column related to gender
df_c3['What is your sex?'] = df_c3['What is your sex?'].apply(convert_gender)

# Displaying the first few rows after transformation
df_c3.head()
```

### 3. Turn all string input to Upper Case

```python
# Function to convert string values to uppercase
def convert_to_uppercase(value):
    if isinstance(value, str):
        return value.upper()
    return value

# Applying the function to all elements in the DataFrame
df_c3 = df_c3.applymap(convert_to_uppercase)

# Displaying the first few rows after transformation
df_c3.head()
```

### 4. Replace unknowns into NaN

```python
# Replacing all occurrences of "UNKNOWN" with NaN
df_c3.replace("UNKNOWN", np.nan, inplace=True)

# Displaying the first few rows after transformation
df_c3.head()
```

### Categorize personality

```python
# Mapping of personality traits to their respective columns
personality_columns = {
    'Neuroticism': ['often_feel_inferior', 'feel_like_going_to_pieces_when_stressed', 'often_feel_tense', 'sometimes_feel_worthless', 'feel_like_giving_up_when_things_go_wrong'],
    'Extraversion': ['enjoy_talking_to_people', 'feel_bursting_with_energy', 'cheerful_high_spirited_person', 'very_active_person'],
    'Openness to experience': ['intrigued_by_art_and_nature', 'try_new_foreign_foods', 'little_interest_in_universe_or_human_condition', 'have_intellectual_curiosity', 'enjoy_playing_with_theories'],
    'Agreeableness': ['get_into_arguments_with_family_or_coworkers', 'seen_as_selfish_or_egotistical', 'seen_as_cold_and_calculating', 'generally_thoughtful_and_considerate'],
    'Conscientiousness': ['keep_belongings_neat_and_clean', 'good_at_pacing_to_get_things_done', 'waste_time_before_settling_to_work', 'sometimes_not_dependable_or_reliable', 'never_able_to_get_organized']
}

# Calculate scores for each personality trait
for trait, columns in personality_columns.items():
    df_c4[trait] = df_c4[columns].sum(axis=1)

# Classify personalities based on the calculated scores
# The maximum score for a trait is simply the product of the scale (1-5) and the number of items/questions for that trait
max_scores = {trait: 5 * len(columns) for trait, columns in personality_columns.items()}

def classify_personality(row):
    # Find the trait with the highest score for the respondent
    return max(max_scores.keys(), key=lambda trait: row[trait])

df_c4['Personality Classification'] = df_c4.apply(classify_personality, axis=1)

# Display the first few rows with the new column
df_c4[['Personality Classification']].head()
```
